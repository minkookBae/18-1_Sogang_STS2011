```python
import re

data = """
ABC01: $23.45 
HGG42: $5.01 
CFXE1: $889.00 
XTC99: $69.89 
Total items found: 4
"""

p = re.compile("(?<=\$).*[.].*")
m = p.findall(data)

for i in m:
    print(i)
```

```python
import re

data = """ 
I paid $30 for 100 apples, 
50 oranges, and 60 pears. 
I saved $5 on this order. 
"""

p1 = re.compile("(?<=\$)\d+")
p2 = re.compile("(?<=\s)\d+(?=\s)")

m1 = p1.findall(data)
print("가격을 나타내는 숫자들")
for i in m1:
    print(i)
print()    
m2 = p2.findall(data)
print("수량을 나타내는 숫자들")
for i in m2:
    print(i)
```
