```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
D = np.sin(X)
plt.plot(X, C,label="cosine")
plt.plot(X, D,label="sin")
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],["$-\pi$","$-\pi/2$","0","$\pi/2$","$\pi$"])
plt.yticks([-1,0,1])
plt.legend(loc=0)
plt.show()
```
