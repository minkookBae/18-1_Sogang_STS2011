```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi,np.pi,256)
S = np.sin(X)
plt.plot(X, S, label = "sin")
plt.scatter([0],[0], color = "r", linewidths=10)
plt.grid(True)
plt.annotate(r"$(0,0)$", xy=(0,0), xycoords="data", xytext=(0.3, 0.7),textcoords="axes fraction", fontsize=16,arrowprops=dict(arrowstyle="->", linewidth=3, color="g"))
plt.legend(loc=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
x = range(10)
y = [v**2 for v in x]
ax1.plot(x,y)
ax2.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1, y1, 'yo-')
plt.title("A table of 2 subplots")
plt.ylabel("Damped oscillation")
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2,"r.-")
plt.xlabel("time (s")
plt.ylabel("Undamped")
print(ax2)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi,np.pi,256)
S = np.cos(X)

dot = 2*np.pi/3

with plt.xkcd():
    plt.plot(X, S, label = "cos")
    plt.title("XKCD style plot!!!")
    plt.yticks([-1.00,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1])
    plt.scatter([dot],[np.cos(dot)], color = "r")
    plt.annotate(r"$0.5 Here!$", xy=(dot,np.cos(dot)), xycoords="data", xytext=(0.5,0.05),textcoords="axes fraction", fontsize=16,arrowprops=dict(arrowstyle="->", linewidth=1, color="k"))
    plt.show()
```
