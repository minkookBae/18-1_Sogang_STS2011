```python
import pandas as pd
import matplotlib.pyplot as plt
raw_data = {
'open' : [11650,11100,11200,11100,11000],
'high' : [12100,11800,11200,11100,11150],
'low' : [11600,11050,10900,10950,10900],
'close' : [11900,11600,11000,11100,11050]
}


df = pd.DataFrame(raw_data, index=["16.02.29","16.02.26","16.02.25","16.02.24","16.02.23"], columns=["open", "high", "low", "close"])
df
```

```python
df.plot(xticks=[0, 1, 2, 3, 4])
plt.title("DaeShin")
plt.show()
```

```python
df.plot(kind="barh")
plt.title("DaeShin")
plt.show()
```

```python
df['close'].plot(xticks=[0,1,2,3,4])
plt.title("Variation of Closing Price")
plt.show()
```
