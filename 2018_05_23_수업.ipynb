```python
%matplotlib inline

from pylab import *

x = linspace(-1.6, 1.6, 10000)
f = lambda x: (sqrt(cos(x)) * cos(200*x) + sqrt(abs(x)) - 0.7 * pow((4-x*x),0.01) )
plot(x, list(map(f,x)))
```

```python
import numpy as np

X = np.linspace(-np.pi,np.pi, 256)
C = np.cos(X)
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])

plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],["$-\pi$","$-\pi/2$","0","$\pi/2$","$\pi$"])
plt.yticks([-1,0,1])
plt.grid(True)
plt.show()
```

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
D = np.sin(X)
plt.plot(X, C,label="cosine")
plt.plot(X, D,label="sin")
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],["$-\pi$","$-\pi/2$","0","$\pi/2$","$\pi$"])
plt.yticks([-1,0,1])
plt.legend(loc=0)
plt.show()
```
